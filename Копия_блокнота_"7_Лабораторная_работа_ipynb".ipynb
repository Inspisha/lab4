# Работа с файлами

Напишите функцию read_first(n_lines, file), которая будет открывать файл file и выводить на печать построчно первые строки в количестве n_lines. Протестируйте функцию на файле [poem.txt](https://disk.yandex.ru/d/FMDR6W_fGTCEkw).



```python
def read_first(n_lines,file):
  f = open (file,"r")
  for _ in range(n_lines):
    line = f.readline()
    print(line,end="")
  f.close()
read_first(4,"poem (1).txt")
```

```python
import pandas as pd
```

```python
df=pd.read_excel("coupon_stats.xlsx")

print("Первые 10 строк данных:")
print(df.head(10))

print("\nКоличество строк и столбцов:")
print(f"Строк:{df.shape[0]}, Столбцов:{df.shape[0]}")

df["coupons_not_used"]=df["coupons_given"]-df["coupons_used"]
df["coupons_not_used"]=df["coupons_not_used"].apply(float)

summary_by_category=df.groupby("good_category")[["coupons_given","coupons_used","coupons_not_used"]].sum()

print("\nСуммарные данные по категориям товаров:")
print(summary_by_category)
```

- любые другие метрики, которые, по вашему мнению, позволят охарактеризовать особенность лексики выбранного текста.

```python
import pandas as pd
import spacy

nlp=spacy.load("en_core_web_sm")

file=open("text.txt","r")
text=file.read()
file.close()

doc=nlp(text)

data=[]
for token in doc:
  if not token.is_punct:
    data.append({
        "token":token.text,
        "lemma":token.lemma_,
        "pos":token.pos_,
        "stop":token.is_stop
    })
df=pd.DataFrame(data)
df.to_excel("text.xlsx")

print("Наиболее часто встречающиеся слова:")
display(df.groupby("lemma").size().nlargest(10).reset_index(name="count"))

print("Наиболее часто встречающиеся части речи:")
display(df.groupby("pos").size().nlargest(10).reset_index(name="count"))

data_stopwords=df.groupby("stop").size()
stopwords=data_stopwords.get(True)
all_words=len(df)
stopwords_per=(stopwords/ all_words)*100
print(f"Процент стоп-слов:{stopwords_per:.2f}%")

not_stop=df[df["stop"]==False]
print("Наиболеевстречающиеся слова, которые не являются стоп-словами:")
display(not_stop.groupby("lemma").size().nlargest(10).reset_index(name="count"))

entities=[(ent.text,ent.label) for ent in doc.ents]
entity_df=pd.DataFrame(entities, columns=["entity","label"])
print("Наиболее часто встречающиеся именованные сущности:")
display(entity_df.groupby("entity").size().nlargest(10).reset_index(name="count"))

entity_token_count= sum(len(ent.text.split()) for ent in doc.ents)
entity_token_per=(entity_token_count/len(doc))*100
print(f"Процент токенов, относящихся к именованным сущностям: {entity_token_per:.2f}%")




```
